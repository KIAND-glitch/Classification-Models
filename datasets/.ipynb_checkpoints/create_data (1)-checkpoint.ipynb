```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv(open("student/student-por.csv"), sep=';')
```

```python
data
```

```python
edumap = {1: 'low', 0:'none', 2:'mid', 3:'mid', 4:'high'}
data['Medu'] = data['Medu'].apply(lambda x: edumap[x])
data['Fedu'] = data['Fedu'].apply(lambda x: edumap[x])



```

```python
timemap = {0:'none', 1: 'low', 2:'medium', 3:'high', 4:'very_high'}
data['traveltime'] = data['traveltime'].apply(lambda x: timemap[x])
data['studytime'] = data['studytime'].apply(lambda x: timemap[x])
data['failures'] = data['failures'].apply(lambda x: timemap[x])
```

```python
relationmap = {1: 'very_bad', 2:'bad', 3:'mediocre', 4:'good', 5:'excellent'}
data['famrel'] = data['famrel'].apply(lambda x: relationmap[x])
data['health'] = data['health'].apply(lambda x: relationmap[x])

```

```python
do_map = {1:'very_low', 2:'low', 3:'mediocre', 4:'high', 5:'very_high'}
data['freetime'] = data['freetime'].apply(lambda x: relationmap[x])
data['goout'] = data['goout'].apply(lambda x: relationmap[x])
data['Dalc'] = data['Dalc'].apply(lambda x: relationmap[x])
data['Walc'] = data['Walc'].apply(lambda x: relationmap[x])



```

```python
def absences(x):
    if x==0:
        return 'none'
    elif 0<x<=3:
        return 'one_to_three'
    elif 3<x<=6:
        return 'four_to_six'
    elif 6<x<=0:
        return 'seven_to_ten'
    return 'more_than_ten'

data['absences'] = data['absences'].apply(lambda x: absences(x))
```

```python
def grade(x):
    if x<10:
        return 'F'
    elif 10<=x<=11:
        return 'D'
    elif 12<=x<=13:
        return 'C'
    elif 14<=x<=15:
        return 'B'
    elif 16<=x<=17:
        return 'A'
    return 'A+'

data['G1'] = data['G1'].apply(lambda x: grade(x))
data['G2'] = data['G2'].apply(lambda x: grade(x))
data['G3'] = data['G3'].apply(lambda x: grade(x))
```
